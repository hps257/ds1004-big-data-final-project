```python
df_bronx_assault_all = pd.read_csv("locationplot/bronx-assault-all.csv", names = ["lat","long"])
df_bronx_assault_all["lat"] = df_bronx_assault_all["lat"].str.replace("(","")
df_bronx_assault_all["long"] = df_bronx_assault_all["long"].str.replace(")","")
df_bronx_assault_all['lat'] = df_bronx_assault_all['lat'].astype(float)
df_bronx_assault_all['long'] = df_bronx_assault_all['long'].astype(float)
# get the coordinates with the most incidents of assault in bronx
new_df =  pd.core.frame.DataFrame({'count' : df_bronx_assault_all.groupby(["lat", "long"] ).size()}).reset_index().sort_values('count',ascending = False)
new_df['index'] = new_df.index
state_geo = r'locationplot/police_precincts.json'

new_df = new_df.head(100)
map = folium.Map(location=[40.8448, -73.8648], zoom_start=13)
new_df.apply(lambda row:folium.Marker(location=[row["lat"], row["long"]]).add_to(map),axis = 1)

map
```
